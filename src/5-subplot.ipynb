### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.linspace(0, 10, 100)
y = x ** 2

bar_categories = ['A', 'B', 'C', 'D']
bar_values = [5, 7, 3, 9]

scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]

pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]
pie_colors = ['blue', 'red', 'green', 'orange']

# Create a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# First subplot: Line chart
axes[0, 0].plot(x, y, color='b', linewidth=2)
axes[0, 0].set_title("Line Chart")
axes[0, 0].set_xlabel("X values")
axes[0, 0].set_ylabel("Y = X")
axes[0, 0].grid(True, linestyle="--", alpha=0.6)

# Second subplot: Bar chart
axes[0, 1].bar(bar_categories, bar_values, color=['red', 'blue', 'green', 'purple'])
axes[0, 1].set_title("Bar Chart")
axes[0, 1].set_xlabel("Categories")
axes[0, 1].set_ylabel("Values")

# Third subplot: Scatter plot
axes[1, 0].scatter(scatter_x, scatter_y, color='blue', marker='o', s=100)
axes[1, 0].set_title("Scatter Plot")
axes[1, 0].set_xlabel("X values")
axes[1, 0].set_ylabel("Y values")
axes[1, 0].grid(True, linestyle="--", alpha=0.6)

# Fourth subplot: Pie chart
axes[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', colors=pie_colors, 
               startangle=140, wedgeprops={'edgecolor': 'black'})
axes[1, 1].set_title("Pie Chart")

# Adjust layout for better display
plt.tight_layout()

# Show the plots
plt.show()
```
